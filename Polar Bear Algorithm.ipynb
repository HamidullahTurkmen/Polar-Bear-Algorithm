```python
import random as rand
import numpy as np
import math
```

```python
def fitness(x, y):
    b = (4 - 2.1 * (x ** 2) + (x ** 4) / 3) * (x ** 2) + x * y + (-4 + 4 * (y ** 2)) * (y ** 2)
    return b
```

```python
T=1000  #iterasyon
k = 100 # ayi sayisi
n=int(k*0.75)  # n %75 ayi
m = 0
r = 0 
s1 = math.pi/2
s2 = math.pi*2
V1 = 0
V2 = 0
best_bear = 0
Xnew = 0
Ynew = 0
current_x = [0.0]*k
current_y = [0.0]*k
current_fitness = [0.0]*k


for i in range(n):
    current_x[i] = rand.uniform(-3, 3)
    current_y[i] = rand.uniform(-2, 2)
    current_fitness[i] = fitness(current_x[i],current_y[i]) 
    

i=0 
while i<T:
    for p in range(n):
        a   = rand.uniform(0,0.3)
        V1 = rand.uniform(0,(s1))            
        V2 = rand.uniform(0,(s2))
       
        r = 4*a*math.cos(V1)*math.sin(V1)
        #denklem 4
        Xnew = current_x[p]+ r*math.cos(V1)
        Ynew = current_y[p]+ r*math.sin(V1)
        New_fitness = fitness(Xnew,Ynew)

        if New_fitness < current_fitness[p]:
            current_x[p] = Xnew
            current_y[p] = Ynew
        else:
            Xnew = current_x[p]- r*math.cos(V1)
            Ynew = current_y[p]- r*math.sin(V1)
            New_fitness = fitness(Xnew,Ynew)
            if New_fitness < current_fitness[p]:
                current_x[p] = Xnew
                current_y[p] = Ynew
    
    
    n2 = int(rand.uniform(0,n*0.1)) # % 10lık bear
    ms, index = (list(t) for t in zip(*sorted(zip(current_fitness, [ind for ind in range(n)]))))
    chosen_index = index[n2]
    # 
    w = math.sqrt((current_x[chosen_index] - current_x[best_bear])**2 + (current_y[chosen_index] - current_y[best_bear])**2)
    r2 = rand.uniform(0,w)
    a2 = rand.uniform(0,1)
    new_x = current_x[chosen_index]+ w*a2+r2
    new_y = current_y[chosen_index]+ w*a2+r2
    New_fitness = fitness(new_x,new_y)
    if New_fitness < current_fitness[chosen_index]:
        current_x[chosen_index] = new_x
        current_y[chosen_index] = new_y
    
    # 
    current_fitness, current_x, current_y = (list(t) for t in zip(*sorted(zip(current_fitness, current_x, current_y))))
    k2 = rand.uniform(0,1)
    if(p < T-1 and k2> 0.75):
        n2 = int(rand.uniform(1,1+n*0.1))  #%10'lik bear
        current_x[n] = (current_x[0]+current_x[n2])/2
        current_y[n] = (current_y[0]+current_y[n2])/2
        current_fitness[n] = fitness(current_x[n],current_y[n])
    elif (n>k/2 and k2<0.25):
        current_x[n] = 0.0
        current_y[n] = 0.0
        current_fitness[n] = 0.0
        n=n-1
    print(current_fitness[n])
    
    i=i+1      
```
